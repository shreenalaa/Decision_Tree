```python
import numpy as np
import pandas as pd
from Decision_Tree import DecisionTreeReg
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("data.csv")
data
```

```python
X = data.drop("y", axis = 1).values
y = data["y"].values
```

```python
model = DecisionTreeReg(max_depth=15)    # overfitting
model.fit(X,y)
```

```python
y_pred = model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred,c = "r")
```

```python
model = DecisionTreeRegressor()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred,c = "r")
```
